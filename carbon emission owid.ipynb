```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#load the dataset
df= pd.read_csv('owid-co2-data.csv')
```

```python
# Filter for a specific country (e.g., 'Nigeria')
country_data = df[df['country'] == 'Nigeria']
```

```python
# Display column names
print(df.columns)
```

```python
# Define the country you're filtering for
country_name = 'Nigeria'
```

```python
df = df.sort_values(by=['country', 'year'])
df
```

```python
# Filter data to include only years from 1750 to 2023
filtered_data = df[(df['year'] >= 1850) & (df['year'] <= 2023)]
```

```python
# Calculate CO2 Emissions per Capita
filtered_data['co2_per_capita'] = filtered_data['co2_including_luc'] / filtered_data['population']
```

```python
# Filter for Nigeria
nigeria_data = filtered_data[filtered_data['country'] == 'Nigeria']
```

```python
# Plot Population and CO2 Emissions (Including LUC) Over Time
plt.figure(figsize=(12, 5))
plt.plot(nigeria_data['year'], nigeria_data['population'], label='Population')
plt.plot(nigeria_data['year'], nigeria_data['co2_including_luc'], label='CO2 Emissions (Including LUC)')
plt.xlabel('Year')
plt.ylabel('Values')
plt.title('Population and CO2 Emissions (Including LUC) Over Time - Nigeria')
plt.legend()
plt.show()
```

```python
# Calculate the correlation between GDP and CO2 Emissions (Including LUC)
correlation = country_data['gdp'].corr(country_data['co2_including_luc'])
print(f'Correlation between GDP and CO2 Emissions (Including LUC) in {country_name}: {correlation}')
```

```python
# Calculate CO2 Emissions per Capita
df['co2_per_capita'] = df['co2_including_luc'] / df['population']
```

```python
# Select the countries you want to compare
selected_countries = ['Nigeria', 'United States', 'China', 'India']
```

```python
# Filter the dataset for the selected countries
comparison_data = df[df['country'].isin(selected_countries)]
comparison_data
```

```python
# Check if the filtered data is not empty
if not comparison_data.empty:
    # Plot CO2 Emissions per Capita for the selected countries
    plt.figure(figsize=(12, 6))
    for country in selected_countries:
        country_data = comparison_data[comparison_data['country'] == country]
        if not country_data.empty:
            plt.plot(country_data['year'], country_data['co2_per_capita'], label=country)
    plt.xlabel('Year')
    plt.ylabel('CO2 Emissions per Capita (tonnes per person)')
    plt.title('CO2 Emissions per Capita Over Time for Selected Countries')
    plt.legend(title='Country')
    plt.show()
else:
    print("No data available for the selectedÂ countries.")
```

```python
# Ensure the data is sorted by year
df = df.sort_values(by=['country', 'year'])
df
```

```python
# Fill missing values using the forward fill method
df['population'] = df['population'].ffill()
df['gdp'] = df['gdp'].ffill()
df['co2_including_luc'] = df['co2_including_luc'].ffill()
```

```python
# Calculate Year-over-Year Growth Rates
df['population_growth'] = df.groupby('country')['population'].pct_change() * 100
df['gdp_growth'] = df.groupby('country')['gdp'].pct_change() * 100
df['co2_growth'] = df.groupby('country')['co2_including_luc'].pct_change() * 100
```

```python
# Display the growth rates for a specific country (e.g., 'Nigeria')
country_name = 'Nigeria'
country_growth = df[df['country'] == country_name][['year', 'population_growth', 'gdp_growth', 'co2_growth']]
print(country_growth)
```
